```python
import sklearn.datasets
import matplotlib.pyplot as plt 
import numpy as np
```

```python
X,Y = sklearn.datasets.make_moons(n_samples=500,noise=2)
```

```python
X.shape
```

```python
Y.shape
```

```python
# X,Y = X.T, Y.reshape(1,Y.shape[0])
Y = Y.reshape(-1,1)
losstrack = []
m = X.shape[0]
w = np.random.randn(1,X.shape[1])*0.01
b = 0
```

```python
X.shape
```

```python
Y.shape
```

```python
w.shape
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
epochs = 1000
learningrate = 0.01
for epoch in range(epochs):
    z = np.dot(X,w.T) + b
    p = sigmoid(z)
    cost = -np.sum(np.multiply(np.log(p),Y) + np.multiply((1-Y),np.log(1-p)))/m
    losstrack.append(np.squeeze(cost))
    dz = p - Y
    dW = (1/m) * np.dot(dz.T,X)
    db = (1/m) * np.sum(dz)
    w = w - learningrate * dW 
    b = b - learningrate * db
    plt.plot(losstrack)
```
